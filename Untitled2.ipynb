```python
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
import os
from skimage import io
```

```python
completeList = os.listdir("visuallyimpair\Visual_dataset/")
for i in completeList:
    temp = io.imread("visuallyimpair\Visual_dataset/" + i)
    temp = resize(temp, (256, 256), anti_aliasing=True)
    io.imsave("visuallyimpair/resize_dataset/{}".format(i), temp)
```
